## 1. Why Incorporate Survey Data into ML Workflows?

Survey data are valuable because they help institutions move beyond *what happened* (outcomes) and toward *why it happened* (experiences and perceptions). In institutional research, that shift is critical when the goal is not just reporting, but **targeted support**.

### 1.1 Survey data capture constructs that administrative data usually miss
Administrative records can tell you **who** left, **when** they left, and **what** their grades were. Survey data can help explain **how students experienced the institution**, such as:
- Academic confidence and preparedness
- Engagement and belonging
- Perceived support and resource use
- Stress, workload, and balance
- Classroom climate and faculty interaction

These constructs often act as early indicators of future outcomes.

### 1.2 Survey data support early warning and intervention
Many institutions want to identify risk *before* the problem appears in GPA or course completion. Survey responses can be used as predictive features in supervised models to help answer questions like:
- Which students may be at risk of stopping out next term?
- Which students may be experiencing high stress that impacts learning?
- Which students may need additional academic support or advising?

### 1.3 Survey-based modeling is often feasible and scalable
Compared to approaches that require costly sensors or complex infrastructure, surveys are typically:
- Lower cost to administer
- Easier to scale across large populations
- Easier to interpret and explain to stakeholders
- More directly connected to intervention strategies

**Key idea:** Surveys can be a practical, institution-friendly entry point into machine learning for student success.

## 2. What “Supervised Learning” Means in a Survey Context

Supervised machine learning means we train a model using **examples with known outcomes**.

### 2.1 The target (y): what are we predicting?
In a survey-based workflow, the outcome might be:
- **Academic success/failure** (binary classification)
- **Stress level** (multiclass classification)
- **GPA** or performance score (regression)
- **Persistence** into a later term (classification)

### 2.2 The features (X): what are we using to predict?
Survey responses become features once they are encoded numerically. Features might include:
- Likert-scale responses (e.g., “Strongly disagree” → 1 … “Strongly agree” → 5)
- Binary yes/no items
- Categorical selections (often one-hot encoded)
- Composite scale scores (averages or sums across items)
- (Later in Course 3) text responses transformed into vectors

### 2.3 Why this matters for institutional research
A supervised model can estimate *risk* or *need*—but the institutional value comes from what you do next:
- Outreach
- Advising intervention
- Resource allocation
- Program design improvement
- Policy or support changes

**Important:** Predictions should be used to support students, not to penalize them.

## 3. Common Supervised Algorithms Used with Survey Data

Survey-based datasets often contain mixed types of variables and complex relationships. In practice, institutions and researchers often compare multiple algorithms.

### 3.1 Algorithm families you’ll see frequently
- **Linear models** (e.g., Logistic Regression): strong baseline, interpretable
- **Tree-based models** (e.g., Decision Trees): intuitive decision rules, handle nonlinearities
- **Ensemble methods** (e.g., Random Forest, Gradient Boosting, XGBoost): often strong performance
- **Kernel methods** (e.g., SVM): effective when boundaries are complex
- **Neural networks** (e.g., MLP/ANN): can learn complex patterns, but may be less interpretable

### 3.2 Why model comparison is common in this space
In survey-based ML, there is rarely a single “best” algorithm for all institutions. Differences in:
- survey instrument,
- student population,
- outcome definition,
- and missingness patterns

can lead to different models performing best. That’s why many studies test several models and choose based on metrics and interpretability.

## 4. Mini Literature Reviews: How Survey Data Were Used in Supervised ML

In this section, we’ll briefly review three published studies that applied supervised ML to survey-derived data in higher education. The articles were all published in *Scientific Reports* in late 2025. Scientific Reports is the 3rd most-cited journal in the world, with more than 834,000 citations in 2024, and receives widespread attention in policy documents and the media. This illustrates the prominent place of higher ed machine learning in academic research and policy.

> **Tip for reading:** As you review each study, focus on four items:
> 1) the outcome they predicted, 2) the survey features they used, 3) the algorithms they tested, and 4) how results were interpreted for action.


### 4.1 Study A: Factors Affecting Academic Performance (Survey + ML)

This study used **survey-collected data** to predict academic performance. The survey included a broad set of variables spanning academic and contextual factors (such as demographics, socioeconomic context, health and well-being indicators, and time-management-related measures). The researchers compared multiple supervised machine learning algorithms and reported that **ensemble methods** performed strongly for prediction. They also emphasized interpretability by examining which factors contributed most to performance prediction, reinforcing the idea that academic outcomes are influenced by both academic history and lived experience.

Marín, Y. R., Quiñones Huatangari, L., Alva Tuesta, J. N., Cruz Caro, O., Maicelo Guevara, J. L., Sánchez Bardales, E., & Chávez Santos, R. (2025). Analysis of factors affecting the academic performance of university students using machine learning. Scientific Reports, 15, Article 44027. https://doi.org/10.1038/s41598-025-28870-1

### 4.2 Study B: Predicting Student Outcomes (Turkey) Using Survey Data

This study framed survey-based prediction as an **early warning system** for student success. Survey variables included student background and perceptions of school context (e.g., socioeconomic indicators, study environment, and related experience factors). The authors compared several supervised classifiers—including **tree-based methods**, **SVM**, and **Random Forest**—and reported that **Random Forest** produced strong, balanced performance after model tuning and validation. A key contribution was translating model results into an intervention-oriented perspective: identifying which survey factors were most associated with the predicted outcome.

Alkan, B. B., Kuzucuk, S., Alkan, N., & Sinan, A. (2025). Using machine learning to predict student outcomes for early intervention and formative assessment. Scientific Reports, 15, Article 39797. https://doi.org/10.1038/s41598-025-23409-w


### 4.3 Study C: Explainable AI for Predictive Modeling of Student Stress (Survey-Based)

This study addressed student well-being by building a **survey-based stress classification model** using a supervised machine learning pipeline. The dataset included features from multiple domains (psychological, physiological, social, environmental, and academic). The authors tested six supervised classification algorithms—**Logistic Regression**, **SVM**, **Decision Tree**, **Random Forest**, **Gradient Boosting**, and **XGBoost**—and reported that **Random Forest** achieved the highest performance (with strong accuracy and balanced classification across stress levels). Importantly, the study used explainability methods (e.g., SHAP) to identify influential predictors such as perceived safety, sleep quality, teacher-student relationship, and extracurricular participation—features that are directly actionable for institutional support planning.

Tariq, R., Orozco-del-Castillo, M. G., Zamir, M. T., Ramírez-Montoya, M. S., & Wilberforce, T. (2025). Explainable artificial intelligence for predictive modeling of student stress in higher education. Scientific Reports, 15, Article 38375. https://doi.org/10.1038/s41598-025-22171-3

## 6. Institutional Implications and Limitations

### 6.1 What these studies suggest for higher ed practice
Across the three examples, survey data are used to:
- Build predictive models that go beyond administrative records
- Identify *actionable* risk and protective factors
- Support targeted interventions (academic support, advising, well-being strategies)

### 6.2 What to watch out for
Survey-based supervised ML also comes with important limitations:
- **Response bias:** students who respond may differ systematically from those who do not
- **Timing effects:** survey responses reflect a moment; experiences can shift quickly across a term
- **Measurement quality:** unclear items or inconsistent scales reduce predictive value
- **Ethical use:** predictions should be used for support, transparency, and care—not surveillance or punishment

A good institutional workflow includes:
- clear governance,
- stakeholder communication,
- and periodic validation of model performance.
