## 1. Why Incorporate Survey Data into ML Workflows?

Survey data are valuable because they help institutions move beyond *what happened* (outcomes) and toward *why it happened* (experiences and perceptions). In institutional research, that shift is critical when the goal is not just reporting, but **targeted support**.

### 1.1 Survey data capture constructs that administrative data usually miss
Administrative records can tell you **who** left, **when** they left, and **what** their grades were. Survey data can help explain **how students experienced the institution**, such as:
- Academic confidence and preparedness
- Engagement and belonging
- Perceived support and resource use
- Stress, workload, and balance
- Classroom climate and faculty interaction

These constructs often act as early indicators of future outcomes.

### 1.2 Survey data support early warning and intervention
Many institutions want to identify risk *before* the problem appears in GPA or course completion. Survey responses can be used as predictive features in supervised models to help answer questions like:
- Which students may be at risk of stopping out next term?
- Which students may be experiencing high stress that impacts learning?
- Which students may need additional academic support or advising?









### 1.3 Survey-based modeling is often feasible and scalable
Compared to approaches that require costly sensors or complex infrastructure, surveys are typically:
- Lower cost to administer
- Easier to scale across large populations
- Easier to interpret and explain to stakeholders
- More directly connected to intervention strategies

**Key idea:** Surveys can be a practical, institution-friendly entry point into machine learning for student success.



## 2. What “Supervised Learning” Means in a Survey Context

Supervised machine learning means we train a model using **labeled data**—examples where the outcomes are already known. By analyzing these examples, the model learns the relationship between student inputs and final results so it can predict outcomes for future students.

### 2.1 The Target ($y$): What are we predicting?
The "Target" is the specific institutional outcome you want to understand. In a survey-based workflow, this typically includes:
* **Academic Success/Failure:** Binary classification (e.g., did the student persist to the next term?).
* **Stress Levels:** Multiclass classification (e.g., Low, Medium, or High stress).
* **Performance Scores:** Regression (e.g., predicting a student's final Term GPA).
* **Persistence:** Classification regarding a student's status in later terms.

### 2.2 The Features ($X$): What are we using to predict?
"Features" are the variables that provide the predictive signal. Survey responses become features once they are encoded numerically:
* **Likert-Scale Responses:** (e.g., “Strongly disagree” → 1 … “Strongly agree” → 5).
* **Binary Items:** Simple Yes/No selections.
* **Categorical Selections:** Often processed via One-Hot Encoding.
* **Composite Scale Scores:** Averages or sums calculated across multiple related items.
* **Text Vectors:** (As seen in previous module) Open-ended responses transformed into PCA components.








### 2.3 Why this matters for Institutional Research
A supervised model can estimate **risk** or **need**, but the true institutional value comes from the intervention that follows:
* **Outreach & Advising:** Triggering proactive support for flagged students.
* **Resource Allocation:** Directing tutoring or mental health services where they are needed most.
* **Program Design:** Improving policies based on the most influential predictors.

<br>

> **Institutional Principle:** Predictions should always be used to **support and care** for students, never to penalize or surveil them.


## 3. Common Supervised Algorithms Used with Survey Data

Survey-based datasets often contain mixed variable types (numeric and categorical) and complex, non-linear relationships. In practice, Institutional Researchers rarely rely on a single model; instead, they compare several "families" of algorithms to find the best balance of accuracy and interpretability.

### 3.1 Algorithm Families You’ll See Frequently
| Algorithm Family | Example | Strength in Institutional Research |
| :--- | :--- | :--- |
| **Linear Models** | [Logistic Regression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html) | The "Gold Standard" for interpretability. Provides clear odds-ratios for each survey item. |
| **Tree-Based Models**| [Decision Trees](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html) | Intuitive "if-then" rules that are easy to explain to non-technical stakeholders. |
| **Ensemble Methods** | [Random Forest](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html) / [XGBoost](https://xgboost.readthedocs.io/) | Generally offers the highest predictive power by combining multiple models to reduce error. |
| **Kernel Methods** | [SVM (Support Vector Machines)](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html) | Effective when the boundaries between student groups (e.g., Stressed vs. Non-Stressed) are highly complex. |
| **Neural Networks** | [MLP (Multi-layer Perceptron)](https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html) | Capable of learning deep, hidden patterns, though they are often seen as "black boxes." |






### 3.2 Why Model Comparison is Vital
There is rarely a single “best” algorithm for all campuses. The "winner" often shifts based on several institutional factors:
* **The Survey Instrument:** The number of items and the scale types (Likert vs. Binary).
* **Population Size:** Some models (like Neural Networks) require massive datasets, while others (like Logistic Regression) thrive on smaller cohorts.
* **Outcome Definition:** Predicting a "Grade" (Regression) requires different math than predicting "Persistence" (Classification).
* **Missing Data Patterns:** How the model handles students who skipped certain questions.

> **Instructor Perspective:** Don't get married to one algorithm. A "Perfect" model in 8.1 might be outperformed by a simpler one in 8.2 once we change the target variable. Always let the data—and the need for explainability—guide your choice.


## 4. Mini Literature Reviews: How Survey Data Were Used in Supervised ML

In this section, we review three studies published in *Scientific Reports* in late 2025. As the 3rd most-cited journal in the world, its focus on these topics highlights the growing importance of machine learning in global education policy.

**As you review each study, focus on four key dimensions:**
1. **The Outcome:** What were they trying to predict?
2. **The Features:** What specific survey items provided the signal?
3. **The Algorithms:** Which models proved most effective?
4. **The Action:** How were the results interpreted for institutional support?


### 4.1 Study A: Predicting Academic Performance
This study moved beyond grades to include the "lived experience" of the student.
* **The Goal:** Predict overall academic performance using a holistic survey.
* **The Features:** Demographics, socioeconomic context, health/well-being indicators, and time-management measures.
* **The Results:** **Ensemble methods** (combining multiple models) performed strongest.
* **The "So What?":** By identifying which non-academic factors contributed most to the prediction, the study reinforced that performance is a product of both history and environment.
> **Citation:** [Marín, Y. R., et al. (2025). Analysis of factors affecting academic performance.](https://doi.org/10.1038/s41598-025-28870-1)

### 4.2 Study B: Early Warning Systems (Turkey)
This research framed survey-based ML as a proactive tool for student retention.
* **The Goal:** Build an early warning system to identify student success vs. failure.
* **The Features:** Background indicators and student perceptions of the school context/environment.
* **The Results:** **Random Forest** produced the most balanced and accurate results after tuning.
* **The "So What?":** The authors translated model outputs into an "intervention-oriented" perspective, helping advisors know which environmental factors were linked to student risk.
> **Citation:** [Alkan, B. B., et al. (2025). Using machine learning to predict student outcomes for early intervention.](https://doi.org/10.1038/s41598-025-23409-w)







### 4.3 Study C: Explainable AI (XAI) for Student Stress
This study is a "Gold Standard" for institutional research because it prioritizes *why* a student is flagged.
* **The Goal:** Classify student stress levels (Low, Medium, High).
* **The Features:** Psychological, physiological, social, environmental, and academic survey domains.
* **The Results:** **Random Forest** achieved the highest accuracy across all stress levels.
* **The "So What?":** Using **SHAP (Explainability)**, researchers identified that *perceived safety* and *sleep quality* were top predictors—factors that are directly actionable for campus well-being planning.
> **Citation:** [Tariq, R., et al. (2025). Explainable artificial intelligence for predictive modeling of student stress.](https://doi.org/10.1038/s41598-025-22171-3)

## 6. Institutional Implications and Limitations

While the studies in Section 4 demonstrate the power of survey-based ML, implementing these models in a real-world campus setting requires careful navigation of technical and ethical boundaries.

### 6.1 Strategic Value for Higher Ed Practice
When executed correctly, these models allow institutions to:
* **Move Beyond Records:** Identify risks that are invisible in traditional GPA or enrollment data.
* **Identify Actionable Factors:** Focus on "modifiable" behaviors (like study habits or belonging) rather than just static demographics.
* **Personalize Support:** Shift from "mass emails" to "targeted outreach" based on specific student needs.

### 6.2 Critical Limitations to Watch Out For
In Institutional Research, "noisy" data is a constant challenge. You must account for:
* **Response Bias:** Students experiencing the highest levels of crisis or disengagement are often the least likely to complete a survey, potentially making them "invisible" to the model.
* **Timing Effects:** A student's outlook in Week 3 (early excitement) is vastly different from Week 12 (finals stress). Models must be context-aware.
* **Measurement Quality:** Vague survey items or inconsistent Likert scales create "statistical noise," which degrades the accuracy of your predictions.








### 6.3 The IR Ethics & Governance Checklist
Before deploying a supervised model on campus, consider these three pillars:
1. **Transparency:** Do students know their feedback is being used for predictive modeling?
2. **Equity & Bias:** Does the model perform equally well for all demographic groups, or does it carry "algorithmic bias"?
3. **The "Care" Mandate:** Are predictions used to trigger **support and intervention**, or are they being used for surveillance and punishment?

<br>

> **Instructor Perspective:** In Course 3, we don't just aim for high "Accuracy Scores." We aim for **Actionable Ethics**. A model that is 99% accurate but results in biased outreach is a failure in the context of student success.
