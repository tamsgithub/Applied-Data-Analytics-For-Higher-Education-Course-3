### How Unsupervised Learning Fits the Course

While Modules 1-3 focused on **supervised learning** (predicting a known outcome like departure), unsupervised learning discovers **hidden patterns** in data without labels. This is valuable for:

- Identifying natural student segments for targeted interventions
- Reducing high-dimensional data for visualization
- Discovering unexpected patterns in student behavior
